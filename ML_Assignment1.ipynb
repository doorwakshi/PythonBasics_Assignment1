**1. What is a parameter?**

Answer:

A parameter is a variable that the model learns during training. It directly affects predictions. For example, weights and biases in neural networks or coefficients in linear regression.

**2. What is correlation? What does negative correlation mean?**

Answer:

Correlation is a statistical measure that shows the extent to which two variables move in relation to each other.

* Positive correlation: both variables increase together.
* Negative correlation: as one variable increases, the other decreases.

**3. Define Machine Learning. What are the main components in Machine Learning?**

Answer:

Machine Learning is a subset of AI where machines learn patterns from data to make decisions without being explicitly programmed.
Main components:

* Dataset
* Model
* Algorithm
* Loss function
* Optimizer
* Evaluation metrics

**4. How does loss value help in determining whether the model is good or not?**

Answer:

The loss value indicates the error in the model's predictions. A lower loss means better performance. High loss means the model is making large errors.

**5. What are continuous and categorical variables?**

Answer:

* Continuous variables: Numeric values with infinite possibilities (e.g., age, salary).
* Categorical variables: Values representing categories or groups (e.g., gender, color).

**6. How do we handle categorical variables in Machine Learning? What are the common techniques?**

Answer:

Common techniques:

* Label Encoding: Assigning a unique number to each category.
* One-Hot Encoding: Creating separate binary columns for each category.

**15. What is causation? Explain difference between correlation and causation with an example.**

Answer:

Causation means one variable directly affects another.

* Correlation: Ice cream sales and drowning increase together.
* Causation: Hot weather causes both, but one doesn't cause the other.

**16. What is an Optimizer? What are different types of optimizers? Explain each with an example.**

Answer:

An optimizer updates model parameters to minimize loss.
Examples:

* SGD: Simple gradient update.
* Adam: Combines momentum and adaptive learning rate.
* RMSprop: Adaptive learning rate using moving average of gradients.
