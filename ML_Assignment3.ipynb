1. What is Simple Linear Regression?

Answer:

Simple Linear Regression is a statistical method that models the relationship between two variables by fitting a linear equation to observed data. One variable is considered independent (X), and the other is dependent (Y).

2. What are the key assumptions of Simple Linear Regression?

Answer:

Linearity between X and Y

Independence of errors

Homoscedasticity (constant variance of errors)

Normal distribution of errors

No multicollinearity (not applicable in simple linear regression but in multiple)

3. What does the coefficient m represent in the equation Y=mx+c?

Answer:

It represents the slope of the line — the change in Y for a one-unit increase in X.

4. What does the intercept c represent in the equation Y=mx+c?

Answer:

It represents the value of Y when X is zero.

6. What is the purpose of the least squares method in Simple Linear Regression?

Answer:

To minimize the sum of the squared differences between observed values and predicted values of the dependent variable.

7. How is the coefficient of determination (R²) interpreted in Simple Linear Regression?

Answer:

R² indicates the proportion of the variance in the dependent variable that is predictable from the independent variable(s).

8. What is Multiple Linear Regression?

Answer:

It is an extension of simple linear regression that uses two or more independent variables to predict a dependent variable.

9. What is the main difference between Simple and Multiple Linear Regression?

Answer:

Simple regression has one independent variable; multiple regression has two or more.

10. What are the key assumptions of Multiple Linear Regression?

Answer:

Same as simple regression with additional:

No multicollinearity among predictors

Linear relationship between predictors and the dependent variable

11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?

Answer:

Heteroscedasticity refers to non-constant variance of errors. It leads to inefficient estimates and invalid significance tests.

12. How can you improve a Multiple Linear Regression model with high multicollinearity?

Answer:

Remove highly correlated predictors

Use Principal Component Analysis (PCA)

Apply ridge regression or lasso

13. What are some common techniques for transforming categorical variables for use in regression models?

Answer:

One-hot encoding

Label encoding

Binary encoding

Ordinal encoding

14. What is the role of interaction terms in Multiple Linear Regression?

Answer:

To model the effect of one variable on the dependent variable that depends on the level of another variable.

15. How can the interpretation of intercept differ between Simple and Multiple Linear Regression?

Answer:

In simple regression, the intercept is the value of Y when X is 0. In multiple regression, it’s the value of Y when all Xs are 0 — which may not always be meaningful.

16. What is the significance of the slope in regression analysis, and how does it affect predictions?

Answer:

It shows the rate of change in the dependent variable with respect to the independent variable. A higher absolute slope implies a stronger influence.

17. How does the intercept in a regression model provide context for the relationship between variables?

Answer:

It anchors the regression line and helps in understanding the expected value of the dependent variable when all predictors are zero.

18. What are the limitations of using R² as a sole measure of model performance?

Answer:

Doesn’t indicate whether the model is appropriate

Can be artificially high with more predictors

Doesn’t show whether predictions are biased

19. How would you interpret a large standard error for a regression coefficient?


Answer:

It indicates high variability and that the coefficient estimate may not be reliable.

20. How can heteroscedasticity be identified in residual plots, and why is it important to address it?

Answer:

A funnel shape or non-random pattern in residual plots suggests heteroscedasticity. It must be addressed to ensure valid inference.

21. What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?

Answer:

It means that some predictors may not be contributing meaningfully to the model and are possibly inflating R² without improving the model.

22. Why is it important to scale variables in Multiple Linear Regression?

Answer:

To ensure all predictors contribute equally and to improve convergence in optimization.

23. What is polynomial regression?

Answer:

It is a type of regression that models the relationship between the dependent and independent variable as an nth degree polynomial.

24. How does polynomial regression differ from linear regression?

Answer:

Polynomial regression fits a nonlinear curve, while linear regression fits a straight line.

25. When is polynomial regression used?

Answer:

When the relationship between variables is nonlinear but still can be approximated using polynomial terms.

27. Can polynomial regression be applied to multiple variables?

Yes, it’s called multivariate polynomial regression.

28. What are the limitations of polynomial regression?


Answer:

Overfitting

Poor extrapolation

Computationally expensive for high-degree polynomials

29. What methods can be used to evaluate model fit when selecting the degree of a polynomial?

Answer:

Cross-validation

Adjusted R²

AIC/BIC

Residual analysis

30. Why is visualization important in polynomial regression?

Answer:

To detect overfitting or underfitting and understand how well the model captures the data pattern.

31. How is polynomial regression implemented in Python?

Answer:

Using numpy and sklearn:




```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
import numpy as np # import numpy for creating sample data

# Create sample data for demonstration
X = np.array([[1], [2], [3], [4], [5]]) # Input features
y = np.array([2, 4, 5, 7, 8])  # Target variable

model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
model.fit(X, y)
```
