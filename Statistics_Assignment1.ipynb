1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.

Answer:

Qualitative Data (Categorical): Describes qualities or characteristics.
Examples: Eye color, gender, nationality.

Quantitative Data (Numerical): Represents measurable quantities.
Examples: Height, weight, temperature.


Scales of Measurement:

Nominal Scale: Categories without a specific order.
Example: Blood types (A, B, AB, O).

Ordinal Scale: Categories with a meaningful order but no consistent difference.
Example: Education level (High school < Bachelor < Master).

Interval Scale: Ordered data with equal intervals, but no true zero.
Example: Temperature in Celsius.

Ratio Scale: Interval scale with a meaningful zero.
Example: Weight (0 kg means no weight).

2. What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.

Answer:

Mean: The average; best for symmetric distributions without outliers.
Example: Average test scores.

Median: Middle value; ideal for skewed data or when outliers exist.
Example: Median household income.

Mode: Most frequent value; useful for categorical data.
Example: Most common shoe size sold.

3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?

Answer:

Dispersion: Describes how spread out data values are.

Variance: Average squared deviation from the mean. Higher variance = more spread.

Standard Deviation: Square root of variance; shows average distance from the mean.
Use: To understand consistency and variability.

4. What is a box plot, and what can it tell you about the distribution of data?

Answer:

A box plot (box-and-whisker plot) is a visual summary of data using five-number summary: minimum, Q1, median, Q3, and maximum.

Shows central tendency, spread, and potential outliers.

Helps in comparing distributions and identifying skewness.

5. Discuss the role of random sampling in making inferences about populations.

Answer:

Random sampling ensures that every individual has an equal chance of selection, reducing bias.

It allows generalization of sample results to the population.

Essential for valid statistical inference.

6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

Answer:

Skewness measures the asymmetry of data.

Positive skew (right): Tail on the right; mean > median.

Negative skew (left): Tail on the left; mean < median.
Skewness affects the choice of central tendency and indicates the direction of outliers.

7. What is the interquartile range (IQR), and how is it used to detect outliers?

Answer:

IQR = Q3 - Q1, the range of the middle 50% of data.

Outliers are values below Q1 - 1.5IQR or above Q3 + 1.5IQR.

IQR helps in robust analysis by minimizing the effect of extreme values.

8. Discuss the conditions under which the binomial distribution is used.

Answer:

Used when:

Fixed number of n trials.

Each trial has only two outcomes: success or failure.

Constant probability of success, p.

Trials are independent.
Example: Tossing a coin 10 times.

9. Explain the properties of the normal distribution and the empirical rule (68–95–99.7 rule).

Answer:

Normal Distribution: Symmetric, bell-shaped curve. Mean = median = mode.
Empirical Rule:

68% of data within 1 SD of mean.

95% within 2 SDs.

99.7% within 3 SDs.
Helps to understand data spread and probability estimates.

10. Provide a real-life example of a Poisson process and calculate the probability for a specific event.

Answer:

Example: Number of customer arrivals at a bank per hour.
If λ = 3 customers/hour, probability of exactly 2 arrivals in one hour:
P(X=2) = (e^–λ * λ^x) / x! = (e^–3 * 3²) / 2! = (0.0498 * 9) / 2 = 0.224.

11. Explain what a random variable is and differentiate between discrete and continuous random variables.

Answer:

A random variable assigns numerical outcomes to random events.

Discrete: Countable outcomes.
Example: Number of heads in 10 tosses.

Continuous: Infinite possible values within a range.
Example: Time taken to run a race.

12. Provide an example dataset, calculate both covariance and correlation, and interpret the results.

Answer:

Dataset:
X = [2, 4, 6], Y = [3, 6, 9]

Covariance:
Mean of X = 4, Mean of Y = 6
Cov(X, Y) = [(2–4)(3–6)+(4–4)(6–6)+(6–4)(9–6)] / 3 = [6 + 0 + 6] / 3 = 4

Correlation (r):
SD(X) = 2, SD(Y) = 3
r = Cov(X,Y) / (SD(X)SD(Y)) = 4 / (23) = 0.667


Interpretation:
Positive covariance and correlation indicate a positive linear relationship between X and Y.
